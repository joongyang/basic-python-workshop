## 통계분석, 기계학습, 심층학습, 빅데이터 ...



* 데이터를 분석(처리)하여 정보(지식)을 추출한다는 점은 공통적이다.

* 데이터를 분석하여 지식을 추출하는 것은 근본적으로 통계학의 영역에 속한다.

* 정보화로 인해 데이터의 규모가 커지고, 컴퓨터 기술의 진보로 인해 전통적인 통계분석 기법을 적용하는 환경이 변했다.

* 데이터 규모가 커짐에 따라서 유의확률, 유의성 검증 등과 같은 통계적 가설 검증은 데이터를 훈련 데이터와 검증 데이터로 분할 하는 것으로 대체되었고,

* 회귀분석, 분산분석, 시계열분석, 판별분석 등과 같은 통계분석법은 자동화되어 깊은 이론적 이해(?)없이도 적용할 수 있게 되었다.

* 기계학습, 심층학습, 빅데이터는 통계학의 새로운 모습이다. 컴퓨터 기술 특히 프로그래밍을 할 수 있어야 하며 대규모 데이터를 관리할 수 있어야 한다.

* 모든 것은 데이터를 분석하는 것이다!!!!

## 자료를 분석하기 위한 소프트웨어

자료를 분석하는 작업 즉 자료의 패턴 또는 지식을 찾는 방법인 통계 분석, 머신러닝 그리고 딥러닝을 구혈할 수 있는 자료분석 소프트웨어는 크게 두종류로 분류할 수 있습니다. 
하나는 상용 소프트웨어이고 하나는 공개 소프트웨어 입니다. 

1. BMDP(Bio-Medical Data Package) : 1965년에 개발, 2017년 이후 제공되지 않음

2. SPSS(Statistical Package for the Social Science) : 1968년 개발, 2009년 IBM에서 인수하여 판매하고 있음

3. SAS(Statistical Analysis System) : 1976년 발표

4. MINITAB : 1972년 개발, 주로 식스시그마 등 품질관리를 위해 사용됨

5. GNU Octave : 수치해석용 자유 컴퓨터 소프트웨어

6. R : 통계 계산과 그래픽을 위한 소프트웨어, 패키지 개발이 용이해서 널리 사용됨

7. Python

최근에는 공개 소프트웨어가 사용 소프트웨어를 압도하고 있으며 이런 추세는 앞으로 지속될 것으로 예상된다.

## 파이썬

**파이썬(Python)**은 2005년부터 2012년까지 구글, 2013년부터 지금까지 Dropbox에서 일하는 네델란드 출신으로 프로그래머 **귀도 반 로섬(Guido van Rossum, 1956~)**이 1991년에 발표한 **고급 프로그래밍 언어**이다.

1. **플랫폼 독립**이며,
2. **인터프리터(interpreter)** 언어이며,
3. **객체지향 프로그래밍(object-oriented programming)**을 지원하며,
4. **동적 타이핑(dynamically typed)**을 지원한다.

파이썬이라는 이름은 귀도가 좋아하는 BBC 코미디 시리즈 *Monty Python's Flying Circus*에서 따온 것이다.
(*Monty Python's Flying Circus*는 코미디 그룹 Monty Python이 제작했으며 1969년부터 1974년 사이에 방영되었다.)

현재 파이썬은 파이썬 소프트웨어 재단이 관리하는 공개 소프트웨어이다.

Note: 프로그램 작성할 때 객체의 타입을 명시적으로 지정하지 않고, 실행시에 자동적으로 결정되록 하는 것을 동적 타이핑이라고 한다.
