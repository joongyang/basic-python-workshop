## 국민건강영양조사
---

1. 국민건강영양조사는 1969년 이래 매년 실시된 국민영양조사와 1962년에 시작된 국민건강 및 보건의식행태조사를 통합한 조사이다.

2. 1998년도에 제1기 조사가 이루어졌으며 3년 주기로 국민의 전반적인 건강과 영양 상태 및 그 추이에 대한 통계 산출을 위해 전국 조사로 진행되고 있다.

3. 국민건강영양조사는 크게 건강설문조사 / 영양조사 / 검진조사의 세 부분으로 나뉘어지며, 과거 한국보건사회연구원, 한국보건산업진흥원, 질병관리본부 등 여러 기관에서 나누어 하던 것을 2007년에 시행한 제4기 조사부터는 질병관리본부에서 통합 시행하고 있다.

4. 전국민을 모집단으로 하며, 최근의 인구주택 총조사를 기본으로 매년 전국 200개 조사구를 선정한다. 각 조사구 당 60여 가구를 대상으로 적합도를 고려한 다음 집단시설가구, 비혈연 6인 이상 가구, 외국인 등으로 구성된 가구, 빈집, 비거주 상가, 노숙 등의 특수 사항을 제외한 23가구를 실제 대상가구로 삼는다. 대상으로 선정된 가구의 구성원은 모두 조사대상이 된다.

### 필요하면 Decoding 작업을 해야 한다.
---

자료 설명서에 보듯이 자료는 원래의 값을 숫자로 바꾼 것이다.
예를 들어, 성별은 남성과 여성 두 가지 값을 가질 수 있다.
자료를 입력할 때 남성은 1, 여성은 2로 입력하였다.
이렇게 실제 값을 숫자 코드로 바꾸는 것을 **부호화(암호화, coding, encoding)**라고 한다.

집계표와 그래프에 숫자 코드가 나오면 결과를 이해하기 힘들 것이다.
집계표와 그래프에 원래의 값이 나오게 하려면 코드와 원래 값을 연결하는 작업을 해야 한다.
즉, 코드를 원래 값으로 환원하는 것은 **복호화(decoding)**라고 한다.

부호화와 복호화를 동시에 가르키는  **codec**이란 용어도 있다.

````python
region_map = {1: "서울", 7: "울산", 13: "전북", 2: "부산", 8: "세종",
              14: "전남", 3: "대구", 9: "경기", 15: "경북", 4: "인천",
              10: "강원", 16: "경남", 5: "광주", 11: "충북", 17: "제주",
              6: "대전", 12: "충남"}                                        # 지역 코드와 실제값을 연결하는 사전 객체를 만든다.

town_map = {1: "동", 2: "읍·면"}                                           # 동/읍면 코드와 실제값을 연결하는 사전 객체를 만든다.

sex_map = {1: "남자", 2: "여자"}                                            # 성별 코드와 실제값을 연결하는 사전 객체를 만든다.
````

코드와 실제값을 연결하는 사전 객체를 이용하여 코드를 실제값으로 변환할 수 있다.
이때 `DataFrame` 객체의 `map()` 메소드를 이용한다.

````python
df["시도"] = df["시도"].map(region_map)    # 사전 객체 region_map을 이용하여 시도 열의 값을 변환한 결과를 시도 열에 저장한다.
df['동읍면'] = df['동읍면'].map(town_map)  # 사전 객체 town_map을 이용하여 동읍면 열의 값을 변환한 결과를 동읍면 열에 저장한다
df['성별'] = df['성별'].map(sex_map)       # 사전 객체 sex_map을 이용하여 성별 열의 값을 변환한 결과를 성별 열에 저장한다
df.head(5)
````

## 범주형 변수에 대한 처리
---

Coding 또는 Decoding을 실시하는 열(변수)은 정해진 몇 가지의 값만 나타난다는 특성을 가지고 있다.
이런 열을 **범주형 변수(categorical variable)**라고 한다.
자료를 올바르게 분석하기 위해서는 해당 열이 범주형 변수임을 지정해야 한다.

특정 열을 범주형 변수로 지정할 때 `pandas.Categorical()` 함수를 이용한다.


````python
df['시도'] = pd.Categorical(df['시도'])      # 시도 열을 범주형으로 지정한다.
df['동읍면'] = pd.Categorical(df['동읍면'])  # 동읍면 열을 범주형으로 지정한다.
df['성별'] = pd.Categorical(df['성별'])      # 성별 열을 범주형으로 지정한다.
df.info()
````

## 집계표 작성
---

일반적으로 집계표는 지역, 성과 같은 범주형 변수의 각 값(범주)에 따라 조사대상자를 여러 집단으로 나눈 다음 각 집단에 대해 다른 변수의 통계치를 계산한 결과를 표로 작성한 것이다.
따라서 행과 열에 배치할 범주형 변수와 통계치를 계산할 변수를 지정해야 한다.
이때 `pivot_table()` 메소드를 이용한다.
`pivot_table()` 메소드의 `index` 매개변수에는 행에 배치할 범주형 변수, `columns` 매개변수에는 열에 배치할 범주형 변수, `values` 매개변수에는 통계치를 계산할 변수를 지정하고, `aggfunc` 매개변수에는 계산할 통계치를 지정한다.

다음 코드는 지역별 성별 연령별 조사대상자의 수를 집계한 표를 작성한다.

````python
df.pivot_table(index=['시도'], columns=['성별'], values = ['개인아이디'],
               aggfunc='count', margins=True, margins_name="합계")              # count는 지료 개수 통계치를 구하라는 지정이다.
````

다음은 시도별, 동읍면별, 성별로 조사대상자의 수를 집계한 표이다.

````python
df.pivot_table(index=['시도', '동읍면'], columns=['성별'], values = ['개인아이디'],
               aggfunc='count', margins=True, margins_name="합계")
````

다음은 시도별 동읍면별 성별로 신장과 체중의 평균을 계산한 집계표이다.

````python
df.pivot_table(index=['시도', '동읍면'], columns=['성별'], values = ['신장', '체중'],
               aggfunc='mean')                                                  # mean은 평균을 구하라는 지정이다.
````

다음은 시도별 동읍면별 성별로 조사대상자수, 신장의 평균, 체중의 평균을 구한 집계표이다.
여러 변수의 다른 통계치를 구할 때는 각 변수에 대해 어떤 통계치를 계산할 것인가를 지정하는 사전 객체를 사용해야 한다.
예를 들어, {'개인아이디': 'count', '신장':'mean', '체중': 'mean'}와 같은 사전 객체를 `aggfunc` 매개변수에 지정하면 된다.

````python
df.pivot_table(index=['시도', '동읍면'], columns=['성별'], values = ['개인아이디', '신장', '체중'],
               aggfunc={'개인아이디': 'count', '신장':'mean', '체중': 'mean'})
````

### 집계표를 excel 파일로 저장할 수 있다.

`pivot_table()` 메소드로 작성한 집계표도 `DataFrame` 객체이며, `to_excel()` 메소드를 이용하면 excel 파일로 저장할 수 있다.

````python
agg_table = df.pivot_table(index=['시도', '동읍면'], columns=['성별'], values = ['개인아이디', '신장', '체중'],
               aggfunc={'개인아이디': 'count', '신장':'mean', '체중': 'mean'})    # 집계표를 agg_table이란 이름으로 저장한다.
agg_table.to_excel("지역별성별집계표.xlsx")                                       # Excel 파일로 저장한다.
````
